# Question 1


**What is the most common start station for all three cities?**

```r
# For New York City

max_count = max(table(ny$Start.Station))
temp = table(ny$Start.Station)
row_count = nrow(temp)
for (id in 1:row_count) {
  if (temp[id] == max_count) {
    print(temp[id])
  } else {
  }
}

ggplot(aes(x = Start.Station), data = ny) +
  geom_bar(stat = 'count')
```

```r
# For Washington

max_count = max(table(wash$Start.Station))
temp = table(wash$Start.Station)
row_count = nrow(temp)
for (id in 1:row_count) {
  if (temp[id] == max_count) {
    print(temp[id])
  } else {
  }
}

ggplot(aes(x = Start.Station), data = wash) +
  geom_histogram(stat = 'count')
```

```r
# For Chicago

max_count = max(table(chi$Start.Station))
temp = table(chi$Start.Station)
row_count = nrow(temp)
for (id in 1:row_count) {
  if (temp[id] == max_count) {
    print(temp[id])
  } else {
  }
}

ggplot(aes(x = Start.Station), data = chi) +
  geom_histogram(stat = 'count')
```

**It can be seen that Washington has the most popular starting station among all three cities, which is the Columbus Circle / Union Station with 1700 trips started from there**

### Question 2

**What is the most common starting month?**

```r
# Total count per starting month for New York City

ny$Date <- as.Date(ny$Start.Time)
ny$Start.Month <- months(ny$Date)

ggplot(data = ny, aes(x = ny$Start.Month)) +
  geom_histogram(stat = 'count') +
  xlab('Starting month') +
  ylab('Total count')
```

```r
# Total count per starting month for Washington

wash$Date <- as.Date(wash$Start.Time)
wash$Start.Month <- months(wash$Date)

ggplot(data = wash, aes(x = wash$Start.Month)) +
  geom_histogram(stat = 'count') +
  xlab('Starting month') +
  ylab('Total count')
```

```r
# Total count per starting month for Chicago

chi$Date <- as.Date(chi$Start.Time)
chi$Start.Month <- months(chi$Date)

ggplot(data = chi, aes(x = chi$Start.Month)) +
  geom_histogram(stat = 'count') +
  xlab('Starting month') +
  ylab('Total count')
```

**As can be seen from the three histograms, the most common starting month for all three cities is June**
